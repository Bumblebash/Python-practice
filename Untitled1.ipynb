```python
numbers = []
Finallist = []
while True:
    inp = input("Enter a number")
    if inp.upper() == "DONE":
        break
    try:
        number = float(inp)
        numbers.append(number)
    except  ValueError:
        print("Enter a valid number")
        continue
for i in numbers:
    if i > 50:
        Finallist.append(i)
print("User Inputs:",numbers)
print("Above 50:", Finallist)
```

```python
numbers = []
while True:
    inp = input("Enter a number: ")
    if inp.upper() == "DONE":
        break
    try:
        number = float(inp)
        numbers.append(number)
    except ValueError:
        print("Enter a valid number")
        continue

# Now check numbers greater than 50
print("Numbers greater than 0:")
for num in numbers:
    if num > 50:
        print(num)
print("All numbers entered:", numbers)
```
