```python
user_input =[]
Finallist =[]
while True:
    inp = input("Enter a number?")
    if inp.lower() == "done":
        break
    try:
        number = float(inp)
        user_input.append(number)   
    except:
        print("ValueError: Enter a valid number")
        continue
for num in user_input:
    if num > 50:
        Finallist.append(num)
print("UserInput:", user_input)
print("Finallist:", Finallist)
    
```
