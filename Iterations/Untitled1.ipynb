```python
numbers = [2,34,5,6,7,89,90]
largest = None
for num in numbers:
    if largest is  None  or num   > largest:
        largest = num
print("largest:", largest)
        
```

```python
numbers = [2,2,2,34,5,6,7,89,90]
smallest = None
for num in numbers:
    if smallest is None or num < smallest:
        smallest = num
print("largest:", smallest)
```
