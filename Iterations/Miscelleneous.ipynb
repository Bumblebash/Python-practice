```python
def min(values):
    smallest = None
    for value in values:
        if smallest is None or value < smallest:
            smallest = value
        return smallest
```

### Largest Number 

```python
numbers  = [3,4,5,78,9,2,23,13,12, 56,78,103]

largest = numbers[0]
for num in numbers:
    if num > largest:
       largest = num
print(largest)
```

### Smallest Number

```python
numbers =  [3,4,5,78,9,2,23,13,12, 56,78,103]
smallest = numbers[0]
for num in numbers:
    if num < smallest:
        num = smallest
print('Smallest', smallest)
```
