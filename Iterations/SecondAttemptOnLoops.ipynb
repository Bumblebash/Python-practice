```python
count = 0
total = 0
while True:
    line = input("Enter a Number:")
    if line.lower() == 'done':
        break
    try:
        number = float(line)
    except:
        print("Value Error: Enter a Numeric value")
        continue 
    count = count + 1
    total = total + number 
average = total / count
print("Count:",count , "Total:",total, "Average:",average)
```

### Get the Maximum Value and Minimum Value from the Input List

### Method I

```python
count = 0 
total = 0
minimum_value = None
maximum_value = None
while True:
    line = input("Enter a Number:")
    if line.lower() == "done":
        break 
    try:
        number = float(line)
    except:
        print("Value Error: Enter a numeric Value")
        continue
    count = count + 1
    total = number + total
    if minimum_value is None or number < minimum_value:
        minimum_value = number
    if maximum_value is None or number >  maximum_value:
        maximum_value = number 
print("total:", total, "count:", count )
print("Min_Value:", minimum_value, "Max_Value:", maximum_value)
    
```
