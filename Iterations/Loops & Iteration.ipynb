## LOOPS & ITERATIONS

```python
n = 5
while n > 0:
    print(n)
    n = n - 1
print('Blastoff!')
```

###  Finishing iterations with 'continue'

```python
while True:
    line = input('> ')
    if line[0] == '#':
        continue
    if line.lower() == 'done':
        break
    print(line)
print('Done!')
```

### Definite Loops

```python
friends  = ['Joseph', 'Glenn', 'Sally']
for friend in friends:
    print('Happy New Year:', friend)
print('Done!')
```

### Maximum and Minimum Loops 

- To find the largest value in a list or seqeence, we construct the following loop:

```python
largest = None 
print('Before:', largest)
for intervar in [3,41,12,9,74,15]:
    if largest is None or itervar > largest:
        largest = intervar 
    print('Loop', intervar, largest )
print('Largest:', largest)
```

```python
values = [3,4,5,6,90,12]
def min(values):
    smallest = None 
    for value in values:
        if smallest is None or value < smallest:
            smallest = value 
    return smallest
    
```

## Exercises
- Write a program which repeatedly reads integers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the integers. If the user enters anything other than an integer, detect their mistake using try and except and print an error message and skip to the next integers

```python
total = 0
count = 0

while True:
    line = input("Enter a number (or 'done' to finish): ")

    if line.lower() == "done":
        break
    try:
        number = float(line)
    except ValueError:
        print("Error: Please enter a numeric value or 'done'. ")
        continue

    total += number
    count += 1

 
    if    count > 0:
        average = total / count 
        print("Total:", total)
        print("Count:", count)
        print("Average", average)
    else:
        print("No numbers were ebtered.")
```

```python
count = 0
total = 0

while True:

    line = input("Enter a number: ")

    if line == "done":
        break

    try:
        number = float(line)
    except:
        print("Invalid input")
        continue

    total = total + number
    count = count + 1

average = total / count

print("Total:", total)
print("Count:", count)
print("Average:", average)
```
