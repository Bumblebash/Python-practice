```python
count  = 0
total = 0
max_value = None
min_value = None
while True:
    line = input("Enter a Value?:")
    if line.lower() == 'done':
        break 
    try:
        number = float(line)
    except:
        print("Invalid Error")
        continue 
    count += 1
    total += number
    average = total / count
    if max_value is None or number > max_value:
        max_value = number
    if min_value is None or number < min_value:
        min_value = number
print("count:", count)
print("total:", total)
print("min_value:", min_value)
print("max_value:", max_value)
print("average:", average)
```
