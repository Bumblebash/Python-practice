# Loops & Interations

- We look at how Python repeats statements using Looping Structures

```python
#Repeated Steps
n = 5
while n > 0:
    print(n)
    n = n - 1
print('Blastoff!')
print(n)
    
```

- Loops (repeated steps) have iteration variables that change each time through a loop. Often these iteratiion variables go through a sequence of numbers.

```python
n = 5 
while n > 0:
    print('Lather')
    print('Rinse')
print('Dry off!')
```

### Breaking Out of a Loop

```python
while True:
    line = input('>')
    if line == 'done' :
        break
    print(line)
print('Done!')
```

# The While Statement

```python
n = 7
while n > 0:
    print(n)
    n = n-1 
print('Hello')
```

## FOR LOOP

```python
friends = ['Joseph', 'Glenn', 'Sally']
for friend in friends:
    print('Happy New Year:', friend)
print('Done!')
```

```python
for friend in friends:
    print('Happy New Year:', friend)
```

## Loop Patterns

- Loops are generally constructed by:
  
       - Initializing one or more variables before the loop starts.
  
       - Perfoming some computation on each item in the loop, possibly changing the variables in the body of the loop.
       - Looping at the resulting variables when the loop completes

## Counting and Summing Loops
- Counting the number of items in a list

```python
count = 0
for interval in [1,2,4,7,8,90, 3,8,9,12]:
    count = count + 1
print('Count:  ', count)
```

```python
total = 0
for intervar in [2,4,5,6,8]:
    total = total + 1
print('Total: ', total)
```

## Maximum and Minimum Loops
- To find the largest value in a list or sequence, we construct the following loop:

```python
largest = None 
print('Before:', largest)
for intervar in [3, 41, 12, 9, 74, 15]:
    if largest is None or intervar > largest:
        largest = intervar
    print('Loop:', intervar, largest)
print('Largest:', largest)
```

### Method 11

```python
numbers  = [3, 41, 12, 9, 74, 15]
largest = numbers[0]  #Start with the first item
for num in numbers:
    if num > largest:
        largest = num 
print('Largest:', largest)
```

### Method III

```python
numbers  = [3,41,12,9,74,15]
largest = min(numbers)
print ('Largest: ', largest)
```
