```python
count = 0
total = 0

while True:

    line = input("Enter a number: ")

    if line == "done":
        break

    try:
        number = float(line)
    except:
        print("Invalid input")
        continue

    total = total + number
    count = count + 1

average = total / count

print("Total:", total)
print("Count:", count)
print("Average:", average)
```
