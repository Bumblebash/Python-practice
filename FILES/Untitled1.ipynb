```python
count = 0
total = 0
# filename = input("mbox-short.txt")
content = open('mbox-short.txt')
for line in content:
    if line.startswith("X-DSPAM-Confidence:"):
        col_pos = line.find(':')
        number = float(line[col_pos + 1:])
        count += 1
        total += number
average = total / count
print(average)
        
    
```
