## Reading through a file 

### 1. Number of Lines in a text file

```python
fhand = open('mbox-short.txt')
count = 0
for line in fhand:
    count += 1
print('Line Count:', count)
```

### 2 . Returning Content in a file 

```python
fhand = open('mbox-short.txt')
content = fhand.read()
print(content)

    
```

#### 3.  Printing Specific Number of Text number from  a file

```python
with open('mbox-short.txt') as file:
    content = file.read()
    print(content[:170])
```

## 4. Number of characters in a text document.

```python
content = open('mbox-short.txt')
inp = content.read()
print(len(inp))
```

# Searching through a file

##### - Using Startswith() Function  to instruct python to start with specific characters

```python
fhand = open('mbox-short.txt')
for line in fhand:
    if line.startswith('From:'):
        continue
        print(line)
```

```python
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
```

```python
fhand = open('mbox-short.txt')
for line in fhand:
    if line.startswith('From:'):
        # continue
        print(line)
```

```python
fname = input('Enter file name:')
try:
    content = open(fname)
except:
    print("Enter a valid file name")
    exit()
count  = 0
for line in content:
    if line.startswith("Subject:"):
        count += 1
        continue 

print("Number of Subjects:", count)
```

```python
fname = input('Enter the file name: ')
try:
    fhand = open(fname)
except:
    print('File cannot be opened:', fname)
    exit()
count = 0
for line in fhand:
    if line.startswith('Subject:'):
        count = count + 1
print('There were', count, 'subject lines in', fname)
```
