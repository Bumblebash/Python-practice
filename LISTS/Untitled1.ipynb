```python
t = [1, 24, 5, 'bash', 'Connrad', 45, 86]

def chop(k):
    if len(k) >= 2:
        del k[0]   # Removes the first element
        del k[-1]  # Removes the last element
    elif len(k) == 1:
        del k[0]   # Empties the list if it only has one item
    # Returns None automatically

# Test the function
result = chop(t)

print("Function returned:", result)  # Output: None
print("Modified list:", t)           # Output: [24, 5, 'bash', 'Connrad', 45]
```
