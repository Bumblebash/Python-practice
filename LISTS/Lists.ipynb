### Lists are Mutable
- Can be changed

```python
numbers = [1, 23, 45, 67, 18]
numbers[3] = "Badbunny"
print(numbers)
```

```python
cheeses = ['Cheeddar', 'Edam', 'Gouda']
print(cheeses[8])
```

### Traversing a List

```python
cheeses = ['Cheeddar', 'Edam', 'Gouda']
for cheese in cheeses:
    print(cheese)
```

```python
numbers = [1,2,3,5,6,7,]
for i  in numbers:
    if i > 2:
        print(i)
```

```python
def get_number_list():
    numbers = []  # 1. Create an empty list
    while True:
        inp = input("Enter a Number? ")
        if inp.lower() == 'done':
            break
        try:
            number = float(inp)
            numbers.append(number)  # 2. Add the valid number to the list
        except ValueError:
            print("Error value, Enter a valid number")
            continue
            for i in number:
                if i > 2:
                    print(number)
    
    return numbers  # 3. Return the full list after the loop breaks

# Example usage:
my_list = get_number_list()
print("Final list:", my_list)
```
