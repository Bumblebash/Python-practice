```python
def get_number_list():
    numbers = [] # 1. Create an empty list
    Final_list = []
    while True:
        inp = input("Enter a Number? ")
        if inp.lower() == "done":
            break
        try:
            number = float(inp)
            numbers.append(number)  # 2. Add the valid number to the list
        except ValueError:
            print("Error value, Enter a valid number")
            continue
    for i in numbers:
        if i > 2:
            Final_list.append(i)
return numbers, Final_List # 3. Return the full list after the loop breaks

# Example usage:
user_input = get_number_list()
print("User_input:", user_input)
print("Final list:", Final_list)
```
