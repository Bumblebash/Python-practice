- If the user enters text that cannot be converted to a number, print "Enter a Valid number" and continue the loop.

𝟒. 𝐃𝐚𝐭𝐚 𝐒𝐭𝐨𝐫𝐚𝐠𝐞:
- Save all valid numerical inputs into a list called "user-input"
- Create a second list called final_list containing only the numbers from the list that are greater than 50 

𝟓. 𝐎𝐮𝐭𝐩𝐮𝐭:
- Return and print the original list of all entries and the filtered list of ages.

```python
def old_people():
    user_input = []
    final_list = []
    while True:
        inp = input("Enter a number?")
        """ Capture users inputs/Numbers until a user types DONE"""
        if inp.upper() == 'DONE':
            break
        try:
            number = float(inp)
            user_input.append(number)
        except ValueError:
            print("Enter a valid number")
            continue
    for age in user_input:
        if age > 50:
            final_list.append(age)
    return user_input, final_list
user_input, final_list = old_people()
print("UserInput:", user_input)
print("Old_people", final_list)
        
        
```
