```python
def loop():
    num = []
    while True:
        inp = input("Enter a number")
        if inp.lower() == "done":
            break
        try:
            number = float(inp)
            num.append(number)
        except ValueError:
            print("Enter a Valid number or Float Value")
    del num[-1]
    del num[0]
    return None
num = loop()
print(num)
        
            
        
```
