```python
import string
fhand = open('romeo.txt')
counts = {}
for line in fhand:
    line = line.translate(str.maketrans('', '', string.punctuation))
    line = line.lower()
    words = line.split()
    for words in words:
        if words not in counts:
            counts[words] = 1
        else:
            counts[words] += 1
#sort the dictionary by values
lst = list()
for key, val in list(counts.items()):
    lst.append((val, key))
lst.sort(reverse=True)
for key, val in lst[:10]:
    print(key, val)
```
