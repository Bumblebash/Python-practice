### Traversal through a String with a Loop

```python
fruit = 'banana'
index = 0
while index < len(fruit):
    letter = fruit[index]
    print(letter)
    index += 1
```

```python
fruit = 'banana'
index = len(fruit) - 1
while index >= 0:
    letter = fruit[index]
    print(letter)
    index -= 1
    
```

```python
fruit = 'banana'
for letter in fruit[::-1]:
    print(letter)
```

### Formatted String Literals

- A formatted string literal (often referred to simply as an f-string) allows Python expressions to be used  within string literals. This is accomplished by prepending an f to the string literal and enclosing expressions in curly braces {}.

```python
camels = 42
f'{camels}'
```

```python
camels = 42 
f'I have spotted {camels} camels'
'I have spotted 42 camels'
```

### Exercise: Slicing strings
Take the following Python code that stores a string:
str = 'X-DSPAM-Confidence: 0.8475'
- Use find and string slicing to extract the portion of the string after the colon character and then use the float function to convert the extracted string into a floating point number.

```python
str =  'X-DSPAM-Confidence:0.8475'
colon_pos = str.find(':')
number_string = str[colon_pos + 1:]
confidence_value = float(number_string)
print(confidence_value)
```

```python
text =  'X-DSPAM-Confidence: 0.8475'
i = 0
while i < len(text):
    try:
        number = float(text[i:])
        break
    except:
        i = i + 1
        continue
print(number)
```
