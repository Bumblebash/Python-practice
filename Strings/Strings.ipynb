### Traversal through a String with a Loop

```python
fruit = 'banana'
index = 0
while index < len(fruit):
    letter = fruit[index]
    print(letter)
    index += 1
```

```python
fruit = 'banana'
index = len(fruit) - 1
while index >= 0:
    letter = fruit[index]
    print(letter)
    index -= 1
    
```

```python
fruit = 'banana'
for letter in fruit[::-1]:
    print(letter)
```

### Looping and Counting

```python
#count letters in tabbii
def (count):
    word = 'tabbii'
    for letter in word:
        if letter == 'b':
            count += 1
print(count)
```
