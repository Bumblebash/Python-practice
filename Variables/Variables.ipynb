2. Write a program to prompt the user for hours and rates per hour to compute gross pay
   - Enter Hours: 35
- Enter Rate: 2.75
- Pay: 96.25

```python
#Convert inputs to float as by default, they are strings
hours  = float(input('Enter Hours:\n'))
rate = float(input('Enter rate:\n'))
#Compute Gross pay
Pay = hours * rate
             
```

```python
#Display the results
print("Pay:", Pay)
```
