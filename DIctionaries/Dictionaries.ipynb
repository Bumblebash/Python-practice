The program creates a Python dictionary that maps the sender's mail address to a count of the number of times they appear in the file. After the dictionary is produced, the program reads through the dictionary using a maximum loop to find the most prolific committer.

```python
#7b
#Most Prolific Commiter
def email_counts():
    while True:
        try:
            file_name = open(input("Enter a file Name"))
            break
        except FileNotFoundError:
            print("Cant open File")
    email_owner = {}
    for line in file_name:
        #Sanity check 
        if not line.startswith("From "):
            continue
        words = line.split()
        emails = words[1]
        email_owner[emails] = email_owner.get(emails, 0) + 1 
        max_emails = None 
        max_count = None
        for email, count in email_owner.items():
            if max_count is None or count > max_count:
                max_emails = email 
                max_count = count
    return email_owner, max_emails, max_count
email_owner, max_emails, max_count = email_counts()
print(max_emails, max_count)
if __name__ == " __main__ ":
    email_counts()
```
